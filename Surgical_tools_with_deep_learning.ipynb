```
dimension=(480, 640, 3)

images01=os.listdir(in_01)
images02=os.listdir(in_02)
images03=os.listdir(in_03)
images04=os.listdir(in_04)

labels = glob.glob('/content/Surgical-Dataset/Labels/label object names/*')

def cropDir(dirFiles, inDir, out):
  for i,image in enumerate(dirFiles):
      bbox=getBoundingBox(image)
      im=Image.open(os.path.join(inDir,image))
      im=im.crop(bbox)
      im.save(out + image, 'JPEG')

def getBoundingBox(imageName):
  with open(labelsPath + imageName[:-4]  + ".txt", 'r') as f:
    read = f.readline()[:-1].split(' ')
    imageClass = int(read[0])
    b_boxes = [float(x) for x in read[1:]]
  
  Y, X = dimension[0], dimension[1]
  x, y, w, h = b_boxes
  x, y, w, h = int((x-w/2) * X), int((y-h/2)*Y), int(w*X), int(h*Y)
  xmin, ymin, xmax, ymax = x, y, x+w, y+h
  return [xmin, ymin, xmax, ymax]

cropDir(images01, in_01, out_01)
cropDir(images02, in_02, out_02)
cropDir(images03, in_03, out_03)
cropDir(images04, in_04, out_04)
    
```

```
target_size=(128, 128)
batch_size = 64

train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    zoom_range=0.05,
    rotation_range=180,
    shear_range=0.1,
    horizontal_flip=True,
    vertical_flip=True)

train_data_dir = "/content/output/train"
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=target_size,
    batch_size=batch_size,
    shuffle=True,
    class_mode='categorical')

validation_datagen = ImageDataGenerator(rescale=1. / 255)
validation_data_dir = "/content/output/val"
validation_generator = validation_datagen.flow_from_directory(
    validation_data_dir,
    target_size=target_size,
    batch_size=1,
    shuffle=True,
    class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1. / 255)
test_data_dir = "/content/output/test"
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=target_size,
    batch_size=1,
    shuffle=False,
    class_mode='categorical')
```
